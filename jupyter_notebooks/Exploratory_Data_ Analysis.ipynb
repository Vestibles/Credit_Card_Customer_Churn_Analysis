# Section 1

## Load libraries and Cleaned Data


```python
# Import necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data

df_encoded = pd.read_csv("Data\cleaned_bank_churn.csv")
```

## 2. Churn Distribution

Section 1 content

```python
churn_counts = df_encoded['Churn'].value_counts()
churn_labels = ['Active', 'Churned']

plt.figure(figsize=(6, 4))
plt.pie(churn_counts, labels=churn_labels, autopct='%1.1f%%', colors=['skyblue', 'salmon'])
plt.title("Churn Distribution")
plt.axis('equal')
plt.show()

# code refined using Copilot
```

## 3. Numerical Features vs. Churn

- a look at how key numeric features differ between churned and active customers.

```python
num_features = [
    'Customer_Age', 'Months_on_book', 'Total_Relationship_Count',
    'Months_Inactive_12_mon', 'Contacts_Count_12_mon',
    'Credit_Limit', 'Avg_Open_To_Buy',
    'Total_Revolving_Bal', 'Total_Trans_Amt', 'Total_Trans_Ct',
    'Total_Trans_Amt_per_Transaction'
]

# Boxplots
plt.figure(figsize=(16, 20))
for i, col in enumerate(num_features):
    plt.subplot(6, 2, i + 1)
    sns.boxplot(x='Churn', y=col, data=df_encoded)
    plt.title(f'{col} by Churn')
plt.tight_layout()
plt.show()

# bug fixed using Copilot
```
